**Another way of doing normalization**

```python
#The goal of normalization is to change the values of numeric columns in the dataset to use a common scale
# without distorting differences in the ranges of values or losing information
data['NumberChildrenAtHome']=data['NumberChildrenAtHome'].apply(lambda x :(x-data.NumberChildrenAtHome.min())/(data.NumberChildrenAtHome.max()-data.NumberChildrenAtHome.min()))
data['TotalChildren']=data['TotalChildren'].apply(lambda x :(x-data.TotalChildren.min())/(data.TotalChildren.max()-data.TotalChildren.min()))
data.head()
```

```python
def studies(x):
  if x=='Bachelors':
        x= 3
  elif x == 'Graduate Degree':
        x = 3
  elif x=='Partial College':
        x= 2 
  elif x=='High School':
        x= 1 
  else:
        x= 0
  return x
```
