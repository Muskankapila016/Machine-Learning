```python
x=['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']
```

```python
y=['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data[x],data[y],test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train
```

```python
x_numpy_array=x_train.to_numpy()
y_numpy_array=y_train.to_numpy()
```

```python
x_transpose=x_numpy_array.transpose()
```

```python
import numpy as np
```

```python
A=np.dot(x_transpose,x_numpy_array)
```

```python
B=np.linalg.inv(A)
```

```python
C=np.dot(x_transpose,y_train)
```

```python
weights=np.dot(B,C)
```

```python
weights
#these are the values of betaa
```

```python
weights.shape
```

```python
x_test.shape
```

```python
prediction=np.dot(x_test,weights)
```

```python
prediction
```

```python
#apply least square method
#def a function for it
def LSM(predicted_values,actual_values):
  A=predicted_values-actual_values
  B=(A**2).sum(axis=0)
  return B
```

```python
LSM(prediction,y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
mymodel=model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores=cross_val_score(mymodel,x_train,y_train,cv=5,scoring='r2')
scores
```

```python
print(np.mean(scores))
```

```python
from sklearn.model_selection import cross_val_predict
predict_test=cross_val_predict(model,x_test,y_test)
predict_test
```

```python
predict_test=cross_val_score(model,x_test,y_test,cv=5)
predict_test
```

```python
print(np.mean(predict_test))
```
